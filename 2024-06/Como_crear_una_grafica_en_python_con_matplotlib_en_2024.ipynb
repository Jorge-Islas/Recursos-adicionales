## Importar librerías

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Gráfica de prueba inicial

El siguiente código ejemplifica el uso de todos los argumentos básicos de la función *plot* de `matplotlib.pyplot`

```python
# Generar datos de prueba
x = np.linspace(0,1,30)
y = np.sin(x)

# Graficar con la función matplotlib.pyplot.plot
plt.plot(
    x, # Coordenadas "x"
    y, # Coordenadas "y"
    c="#000000", # color de línea en formato hexadecimal
    ls="--", # estilo de línea: "-", "--", ".-", ":"
    lw=2, # ancho de línea
    alpha=0.5, # opacidad: 0.5 = 50%, 0.6 = 60%
    marker="D", # marcador para los "puntos": "o", ".", "D", "x", "s", "+"
    mec="#00ff00", # color del borde de los marcadores
    mew=2, # ancho del borde de los marcadores
    mfc="#ffffff", # color de los marcadores
    ms=10, # tamaño de los marcadores
    label="prueba" # etiqueta para lo que graficaste

)
plt.legend() # mostrar la leyenda con las etic¿quetas de lo que grafiques
plt.xlabel("Eje X") # agregar un nombre para el eje "x"
plt.ylabel("Eje Y") # agregar un nombre para el eje "y"
plt.title("Título") # agregar un título al gráfico
plt.savefig("grafica_de_prueba.png", dpi=300) # Guardar gráfica en "grafica_de_prueba.png"
plt.show() # mostrar la gráfica
```

## Ejemplos de gráficas con `matplotlib.pyplot.plot`


### Gráfica de función seno

```python
# Generar datos
x = np.linspace(0, np.pi*2, 100)
y = np.sin(x)

# Graficar función seno
plt.plot(
    x, y,
    ls="-",
    c="r",
    label="seno"
)
# Agregar leyenda, título y nombres a los ejes
plt.title("Gráfica de función seno")
plt.xlabel("Tiempo")
plt.ylabel("Función")
plt.legend()
# Guardar gráfica
plt.savefig("Seno.png", dpi=300)
# Mostrar gráfica
plt.show()
```

### Gráfica de dispersión de dos clusters o grupos de puntos

```python
# Generar datos
from sklearn.datasets import make_blobs
X, y = make_blobs(
    n_samples=100, centers=2, n_features=2,random_state=0, cluster_std=0.5
)

X1 = X[y==0,:]
X2 = X[y==1,:]
```

```python
# Graficar cluster o grupo de puntos 1
plt.plot(
    X1[:,0], X1[:,1],
    marker="D", c="r",
    ms=7, ls="None", mec="k",
    alpha=0.7, label="Cluster 1"
)
# Graficar cluster o grupo de puntos 2
plt.plot(
    X2[:,0], X2[:,1],
    marker="o", c="b",
    ms=7, ls="None", mec="k",
    alpha=0.7, label="Cluster 2"
)
# Agregar leyenda, título y nombres a los ejes
plt.title("Gráfica de Clusters")
plt.xlabel("Característica A")
plt.ylabel("Característica B")
# Guardar gráfica
plt.savefig("Clusters.png", dpi=300)
# Mostrar gráfica
plt.show()
```

### Gráfica de puntos experimentales vs modelo predictivo

```python
# Generar datos
puntos = 15
x = np.linspace(0, 1, puntos)
simular = lambda x, ajuste: x + ajuste * np.random.rand(puntos) - ajuste/2
# Graficar puntos "experimentales"
plt.plot(
    x, simular(x, 0.2),
    ls="None", marker="o",
    alpha=0.7, label="Puntos experimentales",
    mec="#0000ff"
)
# Graficar "modelo matemático"
plt.plot(
    x, x,
    ls="-", c="r",
    marker="D", alpha = 0.8,
    label="Modelo", mec="#ff0000"
)
# Agregar leyenda, título y nombres a los ejes
plt.legend()
plt.xlabel("Tiempo")
plt.ylabel("Valor")
plt.title("Modelo de Regresión Lineal")
# Guardar gráfica
plt.savefig("Regresión.png", dpi=300)
# Mostrar gráfica
plt.show()
```
