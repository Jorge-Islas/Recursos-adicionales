# Medidas de dispersi√≥n

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import iqr
sns.set()
```

```python
beta = 2
valores = np.random.exponential(scale=beta,size=500)
media = beta
varianza = beta**2

sns.histplot(valores)
plt.show()
```

## Rango

```python
rango = np.max(valores) - np.min(valores)

print(f"Rango: {rango:.3f}")
```

## Rango Intercuartil

```python
rango_intercuartil = iqr(valores)

print(f"Rango intercuartil: {rango_intercuartil:.3f}")
fig, ax = plt.subplots(figsize=(9,3))
sns.boxplot(valores, orient="h")
plt.show()
```

## Varianza

```python
varianza = np.var(valores)

print(f"Varianza: {varianza:.3f}")
```

## Desviaci√≥n est√°ndar

```python
desv_estandar = np.std(valores)

print(f"Desviaci√≥n est√°ndar: {desv_estandar:.3f}")
sns.histplot(valores)
plt.axvspan(media - desv_estandar, media + desv_estandar, alpha=0.3, color='red')
plt.axvline(x=media, color='k', lw=2, label="media")
plt.legend()
plt.show()
```
