# Teorema del LÃ­mite Central

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
sns.set()
```

```python
generador = {
    "binomial":    lambda tamanio_muestra: np.random.binomial(10,p=0.2, size=tamanio_muestra),
    "exponencial": lambda tamanio_muestra: np.random.exponential(scale=2,size=tamanio_muestra),
    "gamma":       lambda tamanio_muestra: np.random.gamma(shape=0.5,scale=2,size=tamanio_muestra),
    "poisson":     lambda tamanio_muestra: np.random.poisson(lam=2, size=tamanio_muestra)
}
```

```python
tamanio_muestra = 1000
distribucion = "gamma" # ["binomial","exponencial","gamma","poisson"]

x = generador[distribucion](tamanio_muestra)
sns.histplot(x)
plt.show()
```

```python
num_veces = 10000

# SimulaciÃ³n
medias = np.zeros(num_veces)
for i in range(num_veces):
  medias[i] = np.mean(generador[distribucion](tamanio_muestra))

# Valores de dist. normal
mu = np.mean(medias)
sigma = np.std(medias)
valores_x_normal = np.linspace(np.min(medias),np.max(medias),100)
valores_y_normal = norm.pdf(valores_x_normal, mu, sigma)

# Graficar
sns.histplot(medias, color='r',stat="density")

plt.plot(
    valores_x_normal,
    valores_y_normal,
    color='k',
    lw=2
)

plt.title("$\\bar{x}$"+f"$={mu:.3f},\: S={sigma:.3f}$")

plt.show()
```
