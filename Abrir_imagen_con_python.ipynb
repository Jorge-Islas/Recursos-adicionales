## Abrir imagen con PIL o pillow

```python
with Image.open("image.png") as img:
  display(img)

  imagen_numpy = np.asarray(img)
  print(imagen_numpy.shape)
```

## Abrir imagen con OpenCV

```python
img = imread("image.png")
img = cvtColor(img, COLOR_BGR2RGB) # BGR -> RGB

print(img.shape)
plt.imshow(img)
plt.show()
```
