## Importar librerías

```python
import pandas as pd
```

## Crear DataFrame

```python
datos = pd.DataFrame({
    "tipo":["A","B","A","C","D","C"],
    "categoria":[0,0,1,  2,  3,  2],
    "precio":[0.5,3.0,1.0,2.5,1.25,2.0]
})
datos
```

```python
def mostrar_datos():
  print("Original:")
  display(datos)
  print("-"*28)
  print("Sin duplicados:")
```

## Función `drop_duplicates`

Documentación: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html

### Una columna como criterio

```python
mostrar_datos()

datos.drop_duplicates(
    subset="tipo", keep="first"
)
```

### Múltiples columnas como criterio

```python
mostrar_datos()

datos.drop_duplicates(
    subset=["tipo","categoria"], keep="first"
)
```

### Quitar duplicados excepto última ocurrencia

```python
mostrar_datos()

datos.drop_duplicates(
    subset="tipo", keep="last"
)
```
